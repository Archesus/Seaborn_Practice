```python
import seaborn as sns
import numpy as np
```

```python
penguins = sns.load_dataset("penguins")
```

```python
print(penguins.head())
```

```python
print(penguins["species"].value_counts())
```

```python
print(penguins["island"].value_counts())
```

```python
sns.set_style("ticks")
sns.set_context("notebook")
sns.set_palette("muted")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island")
```

```python
sns.set_style("whitegrid")
sns.set_context("paper")
sns.set_palette("dark")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "species")
```

```python
sns.set_style("dark")
sns.set_context("poster")
sns.set_palette("colorblind")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "sex")
```

```python
sns.set_context("notebook")
sns.set_style("white")
sns.scatterplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", alpha = 0.5)
```

```python
sns.stripplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", jitter=True, dodge=True)
```

```python
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g", hue = "island")
```

```python
sns.histplot(data = penguins, x = "body_mass_g", hue = "species", multiple="stack")
```

```python
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", color="green")
```

```python
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", color="purple", hue = "sex")
```

```python
sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue="sex")
```

```python
sns.barplot(data = penguins, x = "species", y = "flipper_length_mm", hue="sex", estimator=np.median)
```

```python
sns.countplot(data = penguins, x = "species", palette="muted")
```

```python
sns.boxplot(data = penguins, x = "species", y = "body_mass_g")
```

```python
sns.violinplot(data = penguins, x = "species", y = "body_mass_g", hue="sex", palette="muted", split = True)
```

```python
sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", fill= True)
```
